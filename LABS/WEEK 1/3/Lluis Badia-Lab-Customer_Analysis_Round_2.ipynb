## Standarise header names:

```python
# We'll standarise them to lowercase and the spaces we'll replace with '_':

mca_headers = []
for column in range(len(mca.columns)):
    mca_headers.append(mca.columns[column].lower())
```

```python
mca.columns = mca_headers
```

```python
mca.columns
```

```python
# Replacing spaces to '_':
mca.columns = mca.columns.str.replace(' ', '_')
```

```python
# Renaming the 'unnamed:_0' column to 'id':
new_name = {'unnamed:_0': 'id'}
mca.rename(columns = new_name, inplace = True)
```

```python
mca.head(1)
```

## Check and deal with 'NaN' values:

```python
# First of all, we use the function 'dropna()' and then restart the index:
mca = mca.dropna(how = 'all')

# After that, we check the % of NaN has each column:
nulls_df = pd.DataFrame(round(mca.isna().sum() / len(mca), 4) * 100)
```

```python
nulls_df = nulls_df.rename(columns = {'index':'header', 0: 'prop_nulls'})
```

```python
nulls_df
```

```python
# Column 'state' --> we'll replace NaN with 'Unknown'
mca['state'] = mca['state'].fillna("Unknown")
```

```python
# It is necessary to change the name of the variable, otherwise, we'll get an error:
nulls_df2 = pd.DataFrame(round(mca.isna().sum() / len(mca), 4) * 100)
```

```python
nulls_df2 = nulls_df2.rename(columns = {'index':'header', 0: 'prop_nulls'})
nulls_df2
```

```python
# Because 'vehicle_type' has > 50% of NaN values, we'll drop that column:
mca = mca.drop(['vehicle_type'], axis = 1)
```

```python
nulls_df3 = pd.DataFrame(round(mca.isna().sum() / len(mca), 4) * 100)
```

```python
nulls_df3 = nulls_df3.rename(columns = {'index':'header', 0: 'prop_nulls'})
nulls_df3
```

```python
# Now, we'll deal with the numerical columns with NaN:
mslc = mca['months_since_last_claim'].median()
mca['months_since_last_claim'] = mca['months_since_last_claim'].fillna(mslc)

nooc = mca['number_of_open_complaints'].median()
mca['number_of_open_complaints'] = mca['number_of_open_complaints'].fillna(nooc)
mca.head(1)
```

```python
# The rest of categorical columns:
mca['response'] = mca['response'].fillna("Unknown")
mca['vehicle_class'] = mca['vehicle_class'].fillna("Unknown")
mca['vehicle_size'] = mca['vehicle_size'].fillna("Unknown")
```

```python
nulls_df4 = pd.DataFrame(round(mca.isna().sum() / len(mca), 4) * 100)
nulls_df4 = nulls_df4.rename(columns = {'index':'header', 0: 'prop_nulls'})
nulls_df4      # output: no NULL values
```
